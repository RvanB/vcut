```python
import numpy as np
from PIL import Image, ImageOps
from scipy.ndimage import distance_transform_cdt

def compute_cityblock_normals(mask, strength=1.0):
    """
    Generate a normal map from a binary mask using cityblock (L1) distance transform.
    This produces sharper edges with minimal corner rounding.
    """
    # Compute cityblock distance inside the shape (mask)
    dist = distance_transform_cdt(mask, metric='taxicab').astype(np.float32)

    # Compute gradients (central difference approximation)
    dzdx = np.roll(dist, -1, axis=1) - np.roll(dist, 1, axis=1)
    dzdy = np.roll(dist, -1, axis=0) - np.roll(dist, 1, axis=0)

    # Invert direction (simulate carving inward)
    dzdx = -dzdx
    dzdy = -dzdy
    dz = np.full_like(dzdx, 1.0 / strength)

    # Normalize the 3D normals
    norm = np.sqrt(dzdx**2 + dzdy**2 + dz**2 + 1e-8)
    nx = dzdx / norm
    ny = dzdy / norm
    nz = dz / norm

    # Encode to RGB
    normal_rgb = np.zeros((*mask.shape, 3), dtype=np.uint8)
    normal_rgb[..., 0] = ((nx + 1) * 127.5).astype(np.uint8)
    normal_rgb[..., 1] = ((ny + 1) * 127.5).astype(np.uint8)
    normal_rgb[..., 2] = ((nz + 1) * 127.5).astype(np.uint8)

    # Flat normal background
    normal_rgb[~mask] = [128, 128, 255]

    return Image.fromarray(normal_rgb, mode='RGB')

def generate_normal_map_from_image(image_path, output_path="normal_map.png", strength=2.0):
    """
    Loads an image, computes the normal map, and saves the result.
    Assumes black shapes on a white background.
    """
    # Load image and invert (black = shape)
    img = Image.open(image_path).convert("L")
    img = ImageOps.invert(img)

    # Convert to binary mask
    mask = np.array(img) > 128

    # Generate normal map
    normal_map = compute_cityblock_normals(mask, strength=strength)

    # Save and return
    normal_map.save(output_path)
    print(f"Normal map saved to {output_path}")
    return normal_map

# Example usage:
# generate_normal_map_from_image("your_text_image.png", "normal_map.png", strength=2.0)
```
