```python
import numpy as np
from PIL import Image
from scipy.ndimage import distance_transform_cdt
from skimage import measure
from skimage.morphology import binary_erosion

def compute_contour_based_normals(mask, strength=1.0):
    h, w = mask.shape

    contours = measure.find_contours(mask, 0.1)
    contour_img = np.zeros_like(mask)
    for contour in contours:
        for i in range(0, len(contour)):
            y, x = contour[i]
            x = round(x)
            y = round(y)
            contour_img[y, x] = 1

    # Distance transform also gives nearest contour pixel index
    dist, indices = distance_transform_edt(~contour_img, return_indices=True)

    # For each pixel, find vector from nearest contour pixel to current pixel
    yy, xx = np.meshgrid(np.arange(h), np.arange(w), indexing='ij')
    contour_y = indices[0]
    contour_x = indices[1]

    # Vector pointing inward from contour to current pixel
    dx = (xx - contour_x).astype(np.float32)
    dy = -(yy - contour_y).astype(np.float32)

    # Normalize direction vectors
    norm = np.sqrt(dx**2 + dy**2 + 1e-8)
    nx = dx / norm
    ny = dy / norm
    nz = np.full_like(nx, 1.0 / strength)  # simulate height component

    # Normalize the 3D normals
    length = np.sqrt(nx**2 + ny**2 + nz**2)
    nx /= length
    ny /= length
    nz /= length

    # Convert to RGB
    normal_rgb = np.zeros((h, w, 3), dtype=np.uint8)
    normal_rgb[..., 0] = ((nx + 1) * 127.5).astype(np.uint8)
    normal_rgb[..., 1] = ((ny + 1) * 127.5).astype(np.uint8)
    normal_rgb[..., 2] = ((nz + 1) * 127.5).astype(np.uint8)

    # Zero out background
    normal_rgb[~mask] = [128, 128, 255]

    return Image.fromarray(normal_rgb, mode='RGB')
```

```python
normal_map = compute_contour_based_normals(mask, strength=1.0)
normal_map.save("lfa_normal_map.png")
```

```python
from scipy.ndimage import gaussian_filter

# Step 1: Convert RGB normal to XYZ vector in [-1, 1]
normals = (np.array(normal_map).astype(np.float32) / 127.5) - 1.0

# Step 2: Smooth each component independently, masked
smoothed = np.zeros_like(normals)
for i in range(3):
    channel = normals[..., i]
    # Only blur inside the mask
    blurred = gaussian_filter(channel * mask, sigma=1)
    norm_factor = gaussian_filter(mask.astype(np.float32), sigma=1)
    norm_factor[norm_factor == 0] = 1  # avoid divide by zero
    smoothed[..., i] = blurred / norm_factor

# Step 3: Re-normalize the vectors
lengths = np.linalg.norm(smoothed, axis=2, keepdims=True) + 1e-8
normals_smoothed = smoothed / lengths

# Step 4: Convert back to RGB normal map
normal_rgb = ((normals_smoothed + 1.0) * 127.5).astype(np.uint8)
normal_rgb[~mask] = [128, 128, 255]  # flat outside

image = Image.fromarray(normal_rgb)
image.save("norm_something.png")
plt.imshow(normal_rgb)
```
