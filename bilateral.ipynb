```python
import cv2
import matplotlib.pyplot as plt
import numpy

img = cv2.imread('typographic-history-typefaces-origins-normal-1.png')

smoothed = cv2.bilateralFilter(img, d=20, sigmaColor=30, sigmaSpace=75)

img_rgb = cv2.cvtColor(smoothed, cv2.COLOR_BGR2RGB)
plt.figure(dpi=300)
plt.imshow(img_rgb)

cv2.imwrite('typographic-history-typefaces-origins-normal-bilateral.png', smoothed)
```
