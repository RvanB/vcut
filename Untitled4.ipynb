```python
from PIL import Image
import numpy as np
from scipy.ndimage import zoom

def upscale_from_antialiased(img_path, upscale_factor=4, threshold=0.5):
    img = Image.open(img_path).convert("L")
    img_np = np.array(img).astype(np.float32) / 255.0

    # Interpolate to higher resolution (bicubic)
    upscaled = zoom(img_np, upscale_factor, order=3)

    # Threshold to make binary shape
    binary_mask = upscaled >= threshold

    return Image.fromarray((binary_mask * 255).astype(np.uint8))
```
